```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the final dataset
df = pd.read_csv("raw_data/final_dataset.csv")

# Fill any remaining NaNs in return and sentiment columns with 0
df.fillna(0, inplace=True)

# Sentiment columns
sentiment_columns = [
    'statement_sentiment', 'intermeeting_sentiment',
    'MP_Positive_Score', 'MP_Negative_Score',
    'EC_Positive_Score', 'EC_Negative_Score',
    'FG_Positive_Score', 'FG_Negative_Score',
    'BSP_Positive_Score', 'BSP_Negative_Score'
]

# Automatically detect all return columns: T-10 to T+10
returns_columns = sorted(
    [col for col in df.columns if col.startswith('T-') or col.startswith('T+')],
    key=lambda x: int(x.replace('T-', '-').replace('T+', ''))
)

# Summary stats
print("Descriptive statistics for sentiment variables:")
print(df[sentiment_columns].describe())

# Average returns by document type
avg_returns_by_doc = df.groupby("document_type")[returns_columns].mean().T
avg_returns_by_doc.plot(figsize=(12, 6))
plt.title("Average Returns by Document Type (T-10 to T+10)")
plt.xlabel("Days")
plt.ylabel("Average Return")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Correlation matrix between returns and sentiment
correlations = df[returns_columns + sentiment_columns].corr()

# Correlation of returns with sentiment
print("\nCorrelation of returns with statement_sentiment:")
print(correlations['statement_sentiment'][returns_columns])

print("\nCorrelation of returns with intermeeting_sentiment:")
print(correlations['intermeeting_sentiment'][returns_columns])

# Heatmap of sentiment-return correlations
plt.figure(figsize=(12, 6))
sns.heatmap(correlations.loc[returns_columns, sentiment_columns], annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation of Market Returns and Sentiment Scores (T-10 to T+10)")
plt.tight_layout()
plt.show()

# Boxplot of sentiment by document type
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="document_type", y="statement_sentiment")
plt.title("Distribution of Statement Sentiment by Document Type")
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="document_type", y="intermeeting_sentiment")
plt.title("Distribution of Intermeeting Sentiment by Document Type")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load and clean data
df = pd.read_csv("raw_data/final_dataset.csv")
df.fillna(0, inplace=True)
df['announcement_date'] = pd.to_datetime(df['announcement_date'])
df.sort_values('announcement_date', inplace=True)

# Get return columns
returns_columns = sorted(
    [col for col in df.columns if col.startswith('T-') or col.startswith('T+')],
    key=lambda x: int(x.replace('T-', '-').replace('T+', ''))
)

# ------------------------------------
# 1. Correlation Heatmap
# ------------------------------------
sentiment_columns = ['statement_sentiment', 'intermeeting_sentiment']
correlations = df[returns_columns + sentiment_columns].corr()

# Plot heatmap for correlation between sentiment and returns
plt.figure(figsize=(12, 4))
sns.heatmap(correlations.loc[sentiment_columns, returns_columns], annot=True, cmap='coolwarm', center=0)
plt.title("Correlation of GPT Sentiment with Returns (T-10 to T+10)")
plt.xlabel("Return Day")
plt.ylabel("Sentiment Metric")
plt.tight_layout()
plt.show()

# ------------------------------------
# 2. Return by Sentiment Bin (Line Chart)
# ------------------------------------
# Bin the sentiment from -1 to 1 into 10 buckets
df['sentiment_bin'] = pd.cut(df['statement_sentiment'], bins=10)

# Average return per T-day for each sentiment bin
mean_returns_by_bin = df.groupby('sentiment_bin')[returns_columns].mean()

# Plot
plt.figure(figsize=(14, 6))
for t_day in returns_columns:
    plt.plot(mean_returns_by_bin.index.astype(str), mean_returns_by_bin[t_day], label=t_day)

plt.title("Average Return by GPT Statement Sentiment Bin")
plt.xlabel("Statement Sentiment Bin")
plt.ylabel("Average Return")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ------------------------------------
# 3. Heatmap of Statement Sentiment × Return over Time
# ------------------------------------
df_melted = df[['announcement_date', 'statement_sentiment'] + returns_columns].melt(
    id_vars=['announcement_date', 'statement_sentiment'],
    value_vars=returns_columns,
    var_name='T_day',
    value_name='Return'
)

df_melted['SentimentWeightedReturn'] = df_melted['statement_sentiment'] * df_melted['Return']
heatmap_data = df_melted.pivot(index='announcement_date', columns='T_day', values='SentimentWeightedReturn')

plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, cmap='coolwarm', center=0)
plt.title("Statement Sentiment × Return Heatmap by Date (T-10 to T+10)")
plt.xlabel("T-Day")
plt.ylabel("announcement_date")
plt.tight_layout()
plt.show()

# ------------------------------------
# 4. Bullish vs Bearish Performance Comparison
# ------------------------------------
# Classify sentiment
df['sentiment_label'] = df['statement_sentiment'].apply(
    lambda x: 'Bullish' if x > 0.3 else 'Bearish' if x < -0.3 else 'Neutral'
)

avg_returns_by_label = df.groupby('sentiment_label')[returns_columns].mean().T

# Plot
plt.figure(figsize=(12, 6))
for label in ['Bearish', 'Neutral', 'Bullish']:
    plt.plot(avg_returns_by_label.index, avg_returns_by_label[label], label=label)

plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Average Returns by GPT Statement Sentiment Category")
plt.xlabel("T-Day")
plt.ylabel("Average Return")
plt.legend()
plt.tight_layout()
plt.show()
```
